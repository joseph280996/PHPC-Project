```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_scale_result(num_processes, execution_times):
    # Create a figure with a specific size for better visibility
    plt.figure(figsize=(12, 6))

    # Create the main plot
    plt.plot(num_processes, execution_times, 'bo-', linewidth=2, markersize=8)

    # Add a horizontal grid for better readability
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Customize the plot
    plt.title('Parallel Execution Time vs Number of Processes', fontsize=14, pad=15)
    plt.xlabel('Number of Processes', fontsize=12)
    plt.ylabel('Execution Time (ms)', fontsize=12)

    # Set the x-axis to show all integer values
    plt.xticks(num_processes)

    # Add value labels on top of each point

    # Adjust layout to prevent label cutoff
    plt.tight_layout()

    # Show the plot
    plt.show()
```

# Weak Scalability

```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def calculate_sizes(base_size=250, num_procs=40):
    """Calculate image sizes for weak scaling based on base size"""
    sizes = []
    for p in range(1, num_procs + 1):
        new_size = int(base_size * np.sqrt(p))
        sizes.append(new_size)
    return sizes

def resize_and_show_image(input_path, output_dir='../data/scaled_images', base_size=250, num_procs=40):
    """Resize input image and display all versions in notebook"""
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Calculate sizes
    sizes = calculate_sizes(base_size, num_procs)
    
    # Print size information
    print("Weak scaling image sizes:")
    for p, size in enumerate(sizes, 1):
        pixels = size * size
        ratio = pixels / (base_size * base_size)
        print(f"P={p}: {size}x{size} pixels (ratio: {ratio:.2f})")
    
    # Open and resize image
    with Image.open(input_path) as img:
        # Convert to RGB if necessary
        if img.mode != 'RGB':
            img = img.convert('RGB')
            
        # Create figure to show original and all resized images
        # We'll use 4x3 grid (11 images total - original + 10 resized)
        # Create and show all resized versions
        for i, size in enumerate(sizes):
            resized = img.resize((size, size), Image.Resampling.LANCZOS)
            
            # Save image
            output_path = os.path.join(output_dir, f'image_{size}x{size}.png')
            resized.save(output_path)
            
            # Calculate grid position
            row = (i + 1) // 3
            col = (i + 1) % 3
            
            # Display in notebook
        # Turn off any unused subplots
        for i in range(num_procs + 1, 12):
            row = i // 3
            col = i % 3
        
        print("\nAll resized images have been saved to:", output_dir)
        return sizes

# Example usage (run this in a notebook cell):
sizes = resize_and_show_image('../data/input.jpg')
```

```python
def plot_weak_scale_result(processes, execution_times):

    # Create a figure with two subplots sharing the x-axis
    fig, ax1 = plt.subplots(1, 1, figsize=(12, 5), sharex=True)

    # Plot execution times
    ax1.plot(processes, execution_times, 'b-o', linewidth=2, markersize=8)
    ax1.set_title('Weak Scalability Analysis', fontsize=14, pad=20)
    ax1.set_ylabel('Execution Time (seconds)', fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.7)

    # Add image dimension annotations to the first plot

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show the plot
    plt.show()
```
