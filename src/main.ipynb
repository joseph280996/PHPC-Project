# Strong Scalability Test

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_scale_result(num_processes, execution_times):
    # Create a figure with a specific size for better visibility
    plt.figure(figsize=(12, 6))

    # Create the main plot
    plt.plot(num_processes, execution_times, 'bo-', linewidth=2, markersize=8)

    # Add a horizontal grid for better readability
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)

    # Customize the plot
    plt.title('Parallel Execution Time vs Number of Processes', fontsize=14, pad=15)
    plt.xlabel('Number of Processes', fontsize=12)
    plt.ylabel('Execution Time (ms)', fontsize=12)

    # Set the x-axis to show all integer values
    plt.xticks(num_processes)

    # Add value labels on top of each point

    # Adjust layout to prevent label cutoff
    plt.tight_layout()

    # Show the plot
    plt.show()
```

## Rotation

```python
# Data from the parallel execution times
num_processes = np.arange(1, 11)  # 1 to 10 processes
execution_times = np.array([
    5.849,  # 1 process
    3.287,  # 2 processes
    2.492,  # 3 processes
    2.435,  # 4 processes
    2.331,  # 5 processes
    2.668,  # 6 processes
    4.854,  # 7 processes
    5.006,  # 8 processes
    5.089,  # 9 processes
    6.506   # 10 processes
])
plot_scale_result(num_processes, execution_times)
```
